```python
# First, we'll import pandas, a data processing and CSV file I/O library
import pandas as pd

# We'll also import seaborn, a Python graphing library
import warnings # current version of seaborn generates a bunch of warnings that we'll ignore
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)
```

```python
axis_x=np.linspace(-1,2,1000)
```

```python
y1=1-axis_x
```

```python
y1[y1<0]=0
```

```python
y2=axis_x.copy()
y2[y2>0]=0
y2[y2<0]=1
```

```python
axis_x
```

```python
plt.rc('savefig',dpi=400)
plt.plot(axis_x,y1,label="hinge loss")
plt.plot(axis_x,y2,label="0-1 loss")

plt.legend()
```

```python
plt.scatter((2,4,4),(2,4,0),facecolor='blue')
plt.scatter((0,2,0),(0,0,2),facecolor='red')
axis_x=np.linspace(-1,5,1000)
y1=2-axis_x
y2=4-axis_x
sv=3-axis_x

plt.plot(axis_x,y1,linestyle="-.")
plt.plot(axis_x,y2,linestyle="-.")
plt.plot(axis_x,sv)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
```
